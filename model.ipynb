```python
import csv
import numpy as np
def write_csv(fdname):
    with open("./csv.csv", "a") as csv_out_file:
        csv_writer = csv.writer(csv_out_file, delimiter=",", quoting=csv.QUOTE_MINIMAL)
        count = 0
        name = np.array([["x" + str(i), "y" + str(i), "z" + str(i), "visibility" + str(i)] for i in range(33)])
        name = name.flatten().astype(np.str).tolist()
        csv_writer.writerow(["image", "class"] + name)
        for folder in fdname:
            files = os.listdir(folder)
            for image in files:
                img = cv.imread(os.path.join(folder, image))
                if img is not None:
                    img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)
                    with mp_pose.Pose(static_image_mode=False, min_detection_confidence=0.7,  min_tracking_confidence=0.7) as pose:
                        results = pose.process(img_rgb)
                    if results.pose_landmarks is not None:
                        assert len(results.pose_landmarks.landmark) == 33, f"Unexpected no of predicted keypoints{len(results.pose_landmarks.landmark)}"
                        pl = [[lmk.x, lmk.y, lmk.z, lmk.visibility] for lmk in results.pose_landmarks.landmark]
                        pl1 = np.around(pl, 5).flatten().astype(np.str).tolist()
                        if folder == "./not/":
                            csv_writer.writerow([image, "0"] + pl1)
                        else:
                            csv_writer.writerow([image, "1"] + pl1)
                            
                        count+=1
                        
                    else:
                        print(image)
                else: print(image)
            
    print(count)
```
